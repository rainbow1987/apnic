```python
#iplist.txt文件中的行为1个IP地址，循环读取每一个IP（ ip1=line.strip()），调用自定义函数mnt1=getmnt(ip1)
#最后将结果去除后保存于列表mnts和mnts.txt文件中
#已测试 20180405
#20180411整理
from bs4 import BeautifulSoup
import requests
from lxml import etree



#函数功能为输入ip地址，通过该ip地址爬取apnic.net上该ip地址的相关信息，并筛选出该IP的mnt-lower（运营商代号）信息，将该值作为函数返回值。
def getmnt(ip):
    ip_input=str(ip)   #需要查询的ip地址
    payload={"searchtext":ip_input}
    r1=requests.post("https://wq.apnic.net/apnic-bin/whois.pl",data=payload)
    bs=BeautifulSoup(r1.text,"lxml")
    name=bs.title
    tx1=bs.findAll("pre")
    tx2=str(tx1)
    html = etree.HTML(tx2)
    html_data = html.xpath('//*')
    a1=0
    mnt=""
    for i in html_data:
        a1+=1
        if a1==15:
            #print(i.text)
            mnt=i.text
            break

    return mnt


def getiptxt(mntx):
    payload={"searchtext":mntx ,"inverse_attributes":"mnt-lower"  }
    print(mntx)
    r1=requests.post("https://wq.apnic.net/apnic-bin/whois.pl",data=payload)

    bs=BeautifulSoup(r1.text,"lxml")

    name=bs.title

    tx3=bs.findAll("pre")

    tx4=str(tx3)


    with open('Data\\ip\\apnictestv1.txt','a+') as f:
        f.write(tx4)

mnts =[]

with open('Data\\ip\\iplist.txt','r') as ips:
    for line in ips.readlines():
        print( line.strip())
        ip1=line.strip()
        mnt1=getmnt(ip1)
        mnts.append(mnt1) 
mnts = list(set(mnts)) #去重
print(mnts)



n=0
while n<=len(mnts)-1:
    with open('Data\\ip\\mnts.txt','a') as mntsfile:
        mntsfile.write(str(mnts[n])+'\n')
    n+=1
```
